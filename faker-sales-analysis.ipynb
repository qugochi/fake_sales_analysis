```python
from faker import Faker
fake = Faker()
fake_transaction = {
   "Transaction_id": [fake.uuid4() for _ in range (1000)],
    "Customer_name": [fake.name() for _ in range (1000)],
    "Customer_email": [fake.email() for _ in range (1000)],
    "Customer_age": [fake.random_int(min = 18, max= 65) for _ in range (1000)],
    "City": [fake.city() for _ in range (1000)],
    "Product": [fake.company() for _ in range (1000)],
    "Unity_price": [fake.random_int(min= 5000, max= 900000) for _ in range (1000)],
}
df= pd.DataFrame(fake_transaction)
df.to_csv("data_sales.csv", index=False)
```
